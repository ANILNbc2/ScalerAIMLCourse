# Slicing of Array in Pandas

# There are 2 ways of slicing in pandas.
1. iloc : Implicit Index
2. loc : Explicit Index.

iloc[<index>] function is similar to passing the index in an array. <br>
so df.iloc[0] will return the 0th row.<br><br>
for df.iloc[1:10] : <br>
the rows 1-9 will be returned (10 is not included) . Negative indexing will also work with iloc.



Purpose of loc. <br>
The purpose of loc is to do explicit indexing.

df.iloc(0) and we will get the 0th entry<br>

# Duplicates in Data Frames.

1. How do we know if there are any duplicates
2. How do we find the duplicates?

````python
data = {
    "Resturant": ["R1", "R2", "R3","R2", "R4", "R1"],
    "Votes": [120, 240, 150, 240, 80, 120],
    "Rating": [4.1, 4.5, 3.9, 4.5, 4.0, 4.1]
}
sample_df = pd.DataFrame(data)
sample_df
````

````python
sample_df.duplicated()
````

````python
sample_df.duplicated(keep='last')
````

````python
# True only if there are duplicates else False.
sample_df.duplicated(keep=False)
````

````python
# Removes duplicates. This messesup the index
sample_df.drop_duplicates(keep=False)
````

````python
# To fix the index, we will set ignore_index=True or use reset_index(drop=True)
sample_df.drop_duplicates(keep=False, ignore_index=True)
````

````python
sample_df.drop_duplicates(keep='first', ignore_index=True)
````

````python
sample_df.drop_duplicates(keep='last', ignore_index=True)
````
