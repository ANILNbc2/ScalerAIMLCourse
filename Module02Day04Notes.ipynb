#Splitting an array.

````python
import numpy as np
array_1D = np.arange(1,8)
print(array_1D)
#not_working_split = np.split(array_1D, 3)
#This doesnt work, throws error : ValueError: array split does not result in an equal division
split_1d = np.split(array_1D,7)

#This generates a list of arrays :  [array([1]), array([2]), array([3]), array([4]), array([5]), array([6]), array([7])]
print(split_1d)

# Splitting in 2D array :
array_2d = np.arange(1,9).reshape(2,-1)
print(array_2d)
split_2d = np.split(array_2d,2, 0)
print(split_2d)

split_2d = np.split(array_2d,2, 1)
print(split_2d)
````

````python
# Splitting in 2D array :
array_2d = np.arange(1,9).reshape(2,-1)
print(array_2d)
split_2d = np.hsplit(array_2d,2)
print("Horizontal Split : \n")
for arr in split_2d :
  print(arr,"\n")

print("Vertical Split : \n")
split_2d = np.vsplit(array_2d,2)
for arr in split_2d :
  print(arr,"\n")


````

````python
array1 = np.array([[1,2,3]])
array2 = np.array([[10],[11],[12]])
#we cannot do a somple stack as the array shapes are different, but with tile
#we can do :
#array3 = np.tile((array1,array2(3,1)))

#Instead we can do :
# Repeat array1 vertically 3 times
array1_tiled = np.tile(array1, (3, 1))  # shape becomes (3, 3)

#[[ 1  2  3]
# [ 1  2  3]
# [ 1  2  3]]

# Combine side-by-side
#array3 = np.hstack((array1_tiled, array2))
#print(array3)

#Here, by using tile, we are able to stack the array1 multiple times.
array1 = np.array([[1,2,3]])
array1_tiled = np.tile(array1, (2, 1))  # shape becomes (3, 3)
print(array1_tiled)

# For np.tile, we give as params, the number of times we want the array to be
# repeated along the rows and the number of times the cells should be repeated
# along the columns.
# So, for an array [1,2,3], if we give np.tile(array1, (3,2)) then we are saying
# in the final array, the values of the array = (1,2,3) should be repeated 3
# times rowwise and 2 time column wise.

# 3 times row wise :
 # [[1,2,3]
 #  [1,2,3]
 #  [1,2,3]]

 # and 2 times column wise
# [[1,2,3,1,2,3]
#  [1,2,3,1,2,3]
#  [1,2,3,1,2,3]]

# np.tile(array, reps)
# array → The input NumPy array.
# reps → Number of repetitions for each axis:
# If reps is a scalar → repeats along the first axis (for 1D arrays).
# If reps is a tuple → repeats along each axis.
````

#Moving to Pandas.
#Why pandas?

 Labeled Data Support
NumPy arrays are purely positional — you reference data by index.

Pandas adds labels (row indices and column names) so you can reference data more naturally:

python
Copy
import numpy as np
import pandas as pd

arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr, columns=['A', 'B'])

print(df['A'])  # Access by column name instead of position

2️⃣ Handling Heterogeneous Data
NumPy arrays must be homogeneous (all elements same type).

Pandas can store mixed types (int, float, string, datetime) in columns — perfect for real-world datasets.

3️⃣ Built-in Data Cleaning Tools
Pandas has built-in handling for:

Missing data (NaN) with .fillna(), .dropna()

Type conversion (.astype())

Index alignment when combining datasets

NumPy doesn’t have direct missing data handling — you need workarounds.

4️⃣ High-Level Operations
Pandas has groupby, pivot, merge, join, rolling windows — great for analytics.

NumPy focuses on low-level array math — you’d need to code grouping/aggregation manually.

5️⃣ Read/Write Support
Pandas can directly read/write from:

CSV, Excel, SQL, JSON, Parquet, etc.

df = pd.read_csv('data.csv')
NumPy can load text and binary files but has limited format support.

6️⃣ Easier Exploration
Pandas integrates nicely with Jupyter/interactive notebooks:

Pretty table display

.head() / .tail() for quick peeks

Descriptive statistics with .describe()

When to use which
Use Case	Better Tool
Pure numerical computation (linear algebra, ML input arrays)	NumPy
Tabular data, analytics, cleaning, joining	Pandas


Pandas a wrapper over Numpy
